```python
#class structure
import random

class QuizGame:
    def __init__(self):
        self.topics = {
            'World Capitals': [
                {
                    'question': 'What is the capital of Japan?',
                    'options': ['Tokyo', 'Beijing', 'Seoul', 'Bangkok'],
                    'correct_answer': 'Tokyo'
                },
                {
                    'question': 'Can you name the capital of Australia?',
                    'options': ['Sydney', 'Melbourne', 'Canberra', 'Brisbane'],
                    'correct_answer': 'Canberra'
                },
                {
                    'question': 'Which city serves as the capital of Canada?',
                    'options': ['Toronto', 'Ottawa', 'Vancouver', 'Montreal'],
                    'correct_answer': 'Ottawa'
                }
            ],
            'Famous Scientists': [
                {
                    'question': 'Who is known as the father of modern physics and formulated the theory of relativity?',
                    'options': ['Isaac Newton', 'Albert Einstein', 'Galileo Galilei', 'Nikola Tesla'],
                    'correct_answer': 'Albert Einstein'
                },
                {
                    'question': 'Which scientist is credited with the discovery of penicillin?',
                    'options': ['Louis Pasteur', 'Alexander Fleming', 'Marie Curie', 'Robert Koch'],
                    'correct_answer': 'Alexander Fleming'
                },
                {
                    'question': 'Who developed the laws of motion and universal gravitation?',
                    'options': ['Stephen Hawking', 'Isaac Newton', 'Charles Darwin', 'Johannes Kepler'],
                    'correct_answer': 'Isaac Newton'
                }
            ],
            'Classic Literature': [
                {
                    'question': 'Who wrote the play "Romeo and Juliet"?',
                    'options': ['Charles Dickens', 'William Shakespeare', 'Jane Austen', 'Mark Twain'],
                    'correct_answer': 'William Shakespeare'
                },
                {
                    'question': 'In which novel do characters Winston Smith and Julia rebel against a totalitarian regime?',
                    'options': ['1984 by George Orwell', 'Brave New World by Aldous Huxley', 'Fahrenheit 451 by Ray Bradbury', 'The Handmaid\'s Tale by Margaret Atwood'],
                    'correct_answer': '1984 by George Orwell'
                },
                {
                    'question': 'Which novel follows the adventures of a young orphan named Pip and his growth in society?',
                    'options': ['Great Expectations by Charles Dickens', 'Pride and Prejudice by Jane Austen', 'Wuthering Heights by Emily BrontÃ«', 'The Catcher in the Rye by J.D. Salinger'],
                    'correct_answer': 'Great Expectations by Charles Dickens'
                }
            ]
        }
        self.score = 0

#Displaying a Question
     def display_question(self, question):
        print(question['question'])
        for i, option in enumerate(question['options'], start=1):
            print(f"{i}. {option}")
        user_answer = input("Enter the number of your answer: ")
        return user_answer

#Checking the Answer
    def check_answer(self, question, user_answer):
        correct_answer = question['options'].index(question['correct_answer']) + 1
        if user_answer == str(correct_answer):
            print("Correct! Well done.")
            self.score += 1
        else:
            print(f"Sorry, the correct answer is {question['correct_answer']}.")

#Starting the Quiz

    def start_quiz(self):
        print("Welcome to the Quiz Game!")

        # Choose a topic
        print("\nChoose a topic:")
        for i, topic in enumerate(self.topics.keys(), start=1):
            print(f"{i}. {topic}")

        topic_choice = int(input("Enter the number of your chosen topic: "))
        selected_topic = list(self.topics.keys())[topic_choice - 1]
        questions_to_ask = random.sample(self.topics[selected_topic], min(3, len(self.topics[selected_topic])))

        for question in questions_to_ask:
            user_answer = self.display_question(question)
            self.check_answer(question, user_answer)

        print(f"\nQuiz completed! Your score is: {self.score}/{len(questions_to_ask)}")
#Main Execution
if __name__ == "__main__":
    quiz_game = QuizGame()
    quiz_game.start_quiz()
```
